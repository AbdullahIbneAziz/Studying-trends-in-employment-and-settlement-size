### EDA

```python
data['Marital status'].value_counts().plot(kind='pie', autopct='%.2f', figsize=(10,6), colors=['blue', 'orange'])
plt.title('Marital Status Distribution')
plt.show()
```

```python
data['Sex'].value_counts()
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.title('Sex Distribution')
sns.countplot(x=data['Sex'], palette='viridis')
plt.subplot(1,2,2)
plt.title('Education Distribution')
sns.countplot(x=data['Education'])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Sex vs Education')
sns.countplot(x=data['Sex'], hue=data['Education'])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Sex vs Marital Status')
sns.countplot(x=data['Sex'], hue=data['Marital status'], palette='Reds')
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Occupation vs Education')
sns.countplot(x=data['Occupation'], hue=data['Education'])
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Settlement size vs Sex')
sns.countplot(x=data['Settlement size'], hue=data['Sex'], palette='magma')
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Scatterplot Age, Income vs Sex')
sns.scatterplot(x=data['Age'], y=data['Income'], s=100, alpha=1, palette='magma', hue=data['Sex'])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Distribution of Age')
sns.histplot(x=data['Age'], kde=True, color='blue')
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Distribution of Income')
sns.histplot(x=data['Income'], kde=True, color='purple')
plt.show()
```
